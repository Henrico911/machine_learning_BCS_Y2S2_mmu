```python
#Using Google Translate to Translate Basic Statements
```

```python
translator = Translator()
```

```python
text = "Hello Uganda, How are you?"
src_lang = 'en'
dest_lang = 'zh-cn'
```

```python
result = translator.translate(text, src=src_lang, dest=dest_lang)
```

```python
print(f"Translated Text: {result.text}")
```
